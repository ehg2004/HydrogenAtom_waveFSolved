Vamos resolver a equação de Schrödinger para um átomo de Hidrogênio em três dimensões (de forma resumida). Primeiro vamos reescrever a equação de Schorodinger simplificando o fato temporal da equação de onda $\psi{}=\psi{(\vec{r})}$. Reescrevemos da seguinte forma:
(1)$$ -\frac{\hbar^2}{2\mu{}}\nabla{}^2\psi{}+ V(r)\psi{} = E\psi{}  $$ 
Onde a Função de Energia potencial $V=V(\vec{r})$ é dada em função do Energial Potencial Coulombiana, da forma:
(2)$$ V(r) = - \frac{Ze^2}{4\pi\epsilon_{0}r} $$
Onde Z é 1 ( no caso do Hidrogênio);
Inicialmente, reescreveremos a EDP de onda (1) em coordenadas Esfericas, da forma:
$$ x = r\sin{\theta}\cos{{\varphi}} , y = r\sin{\theta}\sin{{\varphi}} , z = r\cos{{\theta}} $$
E reescrevendo o Laplaciano ($\nabla{}^2$) em coordenadas Esféricas, chegamos em:
(3)$$\nabla{}^2 = \frac{1}{r^2}\frac{\partial}{\partial{}r}(r^2\frac{\partial}{\partial{}r}) + \frac{1}{r^2\sin{\theta}^2}\frac{\partial}{\partial{}\theta}(\sin{\theta}\frac{\partial}{\partial{}\theta}) +  \frac{1}{r^2\sin{\theta}^2}\frac{\partial{}^2}{\partial{}\varphi^2}$$
Com $\psi{} = \psi{}(r,\theta,\varphi)$, separamos a dependencia radial da angular da equação de onda da forma:
$$\psi{}(r,\theta,\varphi)= R(r)Y(\theta,\varphi)$$
(Esse será o ponto de partida para ilustrarmos as soluções no fina!!!)
Substituindo (2) e (3) em (1) chegamos em:
$$ -\frac{\hbar^2}{2\mu{}}\left[\frac{Y}{r^2}\frac{\partial}{\partial{}r}(r^2\frac{\partial{}R}{\partial{}r}) + \frac{R}{r^2\sin{\theta}^2}\frac{\partial{}}{\partial{}\theta}(\sin{\theta}\frac{\partial{}Y}{\partial{}\theta}) +  \frac{R}{r^2\sin{\theta}^2}\frac{\partial{}^2}{\partial{Y}\varphi^2}\right]+V(r)RY = ERY $$
Para separar as equações, dividimos por $ -\frac{\hbar^2RY}{2\mu{}r^2}$ e isolamos os termos da forma:
$$ \frac{1}{R}\left[\frac{d}{dr}\left(r^2\frac{dR}{dr} \right) \right] + \frac{2\mu{}r^2}{\hbar^2}\left(E- V(r)\right) = -\frac{1}{Y}\left[ \frac{1}{\sin{\theta}}\frac{\partial{}}{\partial{}\theta}\left(\sin{\theta}\frac{\partial{}Y}{\partial{}\theta}\right) +\frac{1}{\sin{\theta}^2}\frac{\partial{}^2Y}{\partial{}\varphi^2} \right] $$

O lado esquerdo é dependente apenas de $r$ e o lado direito apenas de $\theta$ and $\varphi$ , então ambos os lados serão iguais a uma constante $\lambda_{1}$, que servirá para isolar as soluções: 
$$$$
A equação em R sera a seguinte (agrupando com $\lambda_{1}$ e multiplicando por $\frac{R}{r^2}$):
$$\frac{1}{r^2}\frac{d}{dr}\left(r^2\frac{dR}{dr} \right) + \left \{ \frac{2\mu{}}{\hbar^2}\left(E- V(r)\right) - \frac{\lambda_{1}}{r^2} \right \}R = 0$$
Chega-se em Y agrupando com $\lambda_{1}$ e multiplicando por $Y$:
$$ \frac{1}{\sin{\theta}}\frac{\partial{}}{\partial{}\theta}\left(\sin{\theta}\frac{\partial{}Y}{\partial{}\theta}\right) +\frac{1}{\sin{\theta}^2}\frac{\partial{}^2Y}{\partial{}\varphi^2}  + \lambda_{1}Y = 0$$



A SOLUÇÃO PARA Y: $$ $$
A equação em $Y$ pode ser separada em $Y(\theta,\varphi)=\Theta(\theta)\varPhi(\phi)$, de forma que ,em algumas passagens similares às passagens feitas anteriormente, chegamos nas novas EDOs:
$$\frac{1}{\sin{\theta}}\frac{d}{d\theta}\left( \sin{\theta}\frac{d\Theta}{d\theta} \right) +\left(\lambda_{1} - \frac{\lambda_{2}}{\sin{\theta}^2} \right) = 0 $$
E 
$$\frac{d^2\varPhi}{d\phi^2} + \lambda_{2}\varPhi{} = 0 $$
(Onde $\lambda_{2}$ será o autovalor encontrado de forma análoga a $\lambda_{1}$) $$ $$


A SOLUÇÃO PARA $\varPhi{}(\phi) $: $$ $$
Só é possível encontrar uma solução periodica para $\varPhi$ se $\lambda_{2} \geq 0$. Definiremos um inteiro $m$ como $\lambda_{2}  = m^2$, e encontraremos nossa autofunção:
$$ \varPhi{}_{m}(\phi) = \frac{1}{\sqrt{2\pi{}}}\exp{(im\phi{})} $$ 
 onde já normalizamos $\varPhi{}_{m}$ de forma que:
$$ \int_{0}^{\pi} \langle \varPhi{}_{m},\varPhi{}_{m}  \rangle d\phi{} =1 $$


SOLUÇÂO PARA $\Theta{}(\theta)$ ; $$ $$
Substituiremos a variavel $ \theta{} $ por $w$ tal  que $w=\cos{\theta{}}$ . Dessa forma chegamos em:
$$ \frac{1}{dw}\left[(1-w^2)\frac{dP}{dw} \right]+\left ( \lambda_{1} - \frac{m^2}{1-w^2}\right)P=0$$
Expandindo P em uma série, chegamos que para a série convergir, $\lambda_{1} = l(l+1)$ , onde l é um número inteiro maior que zero. Com inumerosas passagens, chega-se que a autofunção associada a $P_{lm}(w)$ será:
$$P_{lm}(w)=(1-w^2)^{\frac{|m|}{2}}\frac{d^{|m|}P_{l}(w)}{dw^{|m|}}$$
Onde $P_{l}(w) sera o Polinômio de Legendre de grau $l$. Nesse momento, é possivel afirmar que a equção anterior só não é zero se $ l \geq |m|$, portanto restringe-se m.

SOLUÇÂO PARA $R{}(r)$ ; $$ $$
Reescrevendo $R{}(r)$ com $V(r)$ e $\lambda_{1}$, temos:
$$\frac{1}{r^2}\frac{d}{dr}\left(r^2\frac{dR}{dr} \right) + \left \{ \frac{2\mu{}}{\hbar^2}\left(E + \frac{e^2}{4\pi\epsilon_{0}r}\right) - \frac{l(l+1)}{r^2} \right \}R = 0$$
Iremos mudar a variavel da forma:
$$ \rho = \frac{\sqrt{8\mu|E|}}{\hbar}r, u(\rho)=rR(r)$$
Reescrevendo a equação da forma:
$$\frac{d^2u}{d\rho^2}+\left\{\frac{\alpha}{\rho} -\frac{1}{4} -\frac{l(l+1)}{\rho^2} \right\}u=0$$
Com o autovalor $\alpha$:
$$ \alpha = \frac{e^2}{4\pi\epsilon_{0}\hbar}\sqrt{\frac{\mu}{2|E|}}$$
Para valores largos de $\rho$, a equação se simplifica para:
$$\frac{d^2u}{d\rho^2}-\frac{1}{4}u=0$$
Considerando que no infinito, $u(\rho)=0$, a única solução possível é:
$$u(\rho)=\exp{(-\frac{\rho}{2})}$$
Para pequenos valores de $\rho$, podemos multiplicar por $\rho^2$ e desconsiderar os termos lineares e quadraticos que não são multiplicados por derivadas. Dessa forma chegamos na equação:
$$\rho^2\frac{d^2u}{d\rho^2}-l(l+u)=0$$
Que pode ser resolvido por $u = \rho^s$. Cancelando $\rho^s$, chega-se a:
$$s(s-1)-l(l+1)=0$$
Para encontrar uma solução factidica, R(r) tem que tender a uma constante quando r tende a 0. Dessa forma a solução deve ser $s=l+1$.
$$ $$ 
Combinamos essas soluções com uma terceira $F(\rho)$ transitória, temos

$$\frac{d^2F}{d\rho^2} + \left( \frac{2(l+1)}{\rho} -1 \right)\frac{dF}{d\rho} +\frac{\alpha -(l+1)}{\rho}F = 0 $$
Resolvendo com séries, chegamos a conclusão que para a série truncar e convergir, o autovalor $\alpha$ devera ser:
$$ \alpha{} = 1+l+n'$$ 
Onde $n'$ é um numero inteiro positivo. Esse autovalor $\alpha$ será nosso número quântico principal, que nos dará os valores quantizados para a Energia do elétron.
 A solução então para $F(\rho)$ será O polinomio de Laguerre de ordem $ n'$.


SUMARIZANDO AS SOLUÇÕES:
Devemos ainda, normalizar as soluções. Com muito esforço chegaremos em:
$$R_{nlm}(r)=\sqrt{ \left(\frac{2}{a_{0}}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]^3}}\exp{(-\frac{r}{na_{0}})}\left( \frac{2r}{na_{0}}\right)^lL_{n-l-1}(\frac{2r}{na_{0}})$$
e em $m\geq0$:
$$Y_{lm}(\theta,\phi)=(-1)^m\sqrt{\frac{(2l+1)(l-m)!}{4\pi(l+m)!}}P_{lm}(\cos{\theta})\exp{(im\phi)}$$
Nos demais:
$$Y_{lm}(\theta,\phi)=Y^{*}_{l-m}(\theta,\phi)$$

Onde $a_{0}$ é o raio de Bohr,dado por:
$$ a_{0} =\frac{4\pi \epsilon_{0}\hbar^2}{\mu e^2} $$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.gridspec as gridspec
from scipy.special import sph_harm
from scipy.special import laguerre
from scipy.special import gamma
def R(n,l,r):
    Lpoly = laguerre(n-l-1)
    a0=0.05
    Rnorm=np.sqrt(((2/(n*a0))**3)*(gamma(n-l-1+1))/(2*n*(gamma(n+l+1))**3))
    return Rnorm*np.exp(-r)*(2*r)**l*np.polynomial.polynomial.polyval(r,Lpoly)
n=3
r=np.linspace(0,10)

ax1 = plt.subplot2grid((3,3),(0, 0))
ax1.plot(r,R(1,0,r))
ax2 = plt.subplot2grid((3,3),(1 ,0))
ax2.plot(r,R(2,0,r))
ax3 = plt.subplot2grid((3,3),(1 ,1))
ax3.plot(r,R(2,1,r))
ax4 = plt.subplot2grid((3,3),(2, 0))
ax4.plot(r,R(3,0,r))
ax5 = plt.subplot2grid((3,3),(2, 1))
ax5.plot(r,R(3,1,r))
ax6 = plt.subplot2grid((3,3),(2, 2))
ax6.plot(r,R(3,2,r))
plt.show()
```

```python
from scipy.special import sph_harm

plt.rc('text', usetex=True)

# extraido de https://scipython.com/blog/visualizing-the-real-forms-of-the-spherical-harmonics/

theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)

theta, phi = np.meshgrid(theta, phi)

def plot_Y(ax, l, m):
    Y = sph_harm(abs(m), l, phi, theta)

    # separates "clouds"
    if m < 0:
        Y = Y.imag
    elif m > 0:
        Y = Y.real
    # tente por voce mesmo:
    #Y =Y*np.conjugate(Y)
    # 
    # 

    # cartesian coordinates - plota em função de theta e phi e usa a função de onda como o raio
    Yx = np.abs(Y) * np.sin(theta) * np.cos(phi)
    Yy = np.abs(Y) * np.sin(theta) * np.sin(phi)
    Yz = np.abs(Y) * np.cos(theta)

    # color mapping (sign of Y)
    cmap = plt.cm.ScalarMappable(cmap=plt.get_cmap('PRGn'))
    cmap.set_clim(-0.5, 0.5)

    ax.plot_surface(Yx, Yy, Yz,
                    facecolors=cmap.to_rgba(Y.real),
                    rstride=2, cstride=2)

    # Axes for visualization
    ax_lim = 0.5
    ax.plot([-ax_lim, ax_lim], [0,0], [0,0], c='0.5', lw=1, zorder=10)
    ax.plot([0,0], [-ax_lim, ax_lim], [0,0], c='0.5', lw=1, zorder=10)
    ax.plot([0,0], [0,0], [-ax_lim, ax_lim], c='0.5', lw=1, zorder=10)

    # ax.set_title(r'$Y_{{{},{}}}$'.format(l, m))
    ax_lim = 0.5
    ax.set_xlim(-ax_lim, ax_lim)
    ax.set_ylim(-ax_lim, ax_lim)
    ax.set_zlim(-ax_lim, ax_lim)
    ax.axis('off')

l_max = 3
figsize_px, DPI = 800, 100
figsize_in = figsize_px / DPI
fig = plt.figure(figsize=(figsize_in, figsize_in), dpi=DPI)
#fig = plt.figure()

spec = gridspec.GridSpec(ncols=2*l_max+1, nrows=l_max+1, figure=fig)
for l in range(l_max+1):
    for m_l in range(-l, l+1):
        ax = fig.add_subplot(spec[l, m_l+l_max], projection='3d')
        plot_Y(ax, l, m_l)
        #retirei o titulo devido a conflito de versoes-codigo nao rodava
plt.show()
```
